```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
r1 = requests.get('https://www.agjeans.com/women/clothing/denim')
soup = bs(r1.content, 'html.parser')
print(soup.prettify())
```

```python
print(r1.text)
```

```python
# Main logic of the items

start = 0
new_link = []
newdesc = []
new_price = []
new_title = []

while True:
    
    r1 = requests.get(f'https://www.agjeans.com/women/clothing/denim?start={start}&sz=36&format=page-element')
    print(r1.status_code)
    soup1 = bs(r1.content, 'html.parser')
    lst = []
    pick=soup1.find_all('div', class_="product-information")
    
    description = soup.find_all('div',class_='product-description')
    descloop = [desc.text.strip() for desc in description]
    
    price = soup.find_all('div',class_='product-pricing')
    priceloop = [p.text.strip() for p in price]
    
    title = soup.find_all('a',class_='name-link')
    titleloop = [titles.text.strip() for titles in title]
    
    for link in soup1.find_all('a',class_='thumb-link'):
        lst.append(link.get('href'))
    if lst == [] or descloop ==[] or priceloop == [] or titleloop == []:
        break
    new_link.extend(lst)
    newdesc.extend(descloop)
    new_price.extend(priceloop)
    new_title.extend(titleloop)
    start+=36
```

```python
retailer = []
for i in range(280):
    retailer.append("agjeans")
len(retailer)
```

```python
# Fetching the complete url of the each items

base_url = "https://www.agjeans.com"
url_p = []
anchors = soup.find('a',class_='name-link')
for i in range(len(new_link)):
    url_p.append(base_url+new_link[i])
```

```python
# Finding SKU of the each items

url = []
for i in url_p:
    r2 = requests.get(i)
    print(r2.status_code)
    soup = bs(r2.content, 'html.parser')
    url.append(soup.find('div',class_='product-number').span.text)
print(url)
```

```python
# Formatting the prices of the each items.

new_price2 = []
for i in new_price:
    new_price2.append(i.lstrip("₹"))
new_price2
```

```python
# Availabilty [size of the each items]

size = []
size_li = []
for i in url_p:
    r3 = requests.get(i)
    print(r3.status_code)
    soup = bs(r3.content, 'html.parser')
    main_div = soup.find_all('li',class_='variation-group-value')
    li = []
    for j in main_div:
        main_li = j.text.strip()
        li.append(main_li)
    size_li.append(li)

print(size_li)
```

```python
import pandas as pd
nData = {
    'Retailer' : retailer,
    'Title' : new_title,
    'Product': newdesc,
    'Price': new_price2,
    'Available':size_li,
    'SKU' : url,
    'Website' : url_p,
}
```

```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
webpage = requests.get('https://www.banggood.in/search/game-console.html')

soup = bs(webpage.content,'html.parser')
print(soup)
```

```python
print(soup.prettify())
```

```python
title = soup.find_all('a',class_='title')
sell_price = soup.find_all('span',class_ = 'price')
review = soup.find_all('a',class_ = 'review')
```

```python
titleloop = [titles.text for titles in title ]
selloop = [sells.text for sells in sell_price]
reviewloop = [reviews.text for reviews in review ]
```

```python
for i in titleloop:
    print(i)
```

```python
data = {
    'Name_of_console':titleloop,
    'Selling_price':selloop,
    'No_of_Review': reviewloop,
}
```

```python
print(len(titleloop),len(selloop),len(reviewloop))
```

```python
# AS the length of the each section is not equal so we have to maintain
# separate data frame for equal length of the list.
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(data,columns=[
    'Name_of_console',
    'Selling_price',
])
```

```python
df2 = pd.DataFrame(data,columns=[
    'No_of_Review',
])
```

```python
# join the both data frames
```

```python
No_of_Reviews = df2['No_of_Review']
```

```python
df1 = df1.join(No_of_Reviews)
```

```python
start = 0
new_link = []
newdesc = []
new_price = []
new_title = []

while True:
    
    r1 = requests.get(f'https://www.agjeans.com/women/clothing/denim?start={start}&sz=36&format=page-element')
    print(r1.status_code)
    soup1 = bs(r1.content, 'html.parser')
    lst = []
    pick=soup1.find_all('div', class_="product-information")
    
    description = soup.find_all('div',class_='product-description')
    descloop = [desc.text.strip() for desc in description]
    
    price = soup.find_all('div',class_='product-pricing')
    priceloop = [p.text.strip() for p in price]
    
    title = soup.find_all('a',class_='name-link')
    titleloop = [titles.text.strip() for titles in title]
    
    for link in soup1.find_all('a',class_='thumb-link'):
        lst.append(link.get('href'))
    if lst == [] or descloop ==[] or priceloop == [] or titleloop == []:
        break
    new_link.extend(lst)
    newdesc.extend(descloop)
    new_price.extend(priceloop)
    new_title.extend(titleloop)
    start+=36
```

```python
base_url = "https://www.agjeans.com"
url_p = []
anchors = soup.find('a',class_='name-link')
for i in range(len(new_link)):
    url_p.append(base_url+new_link[i])
len(url_p)
```

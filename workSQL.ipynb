```python
import pandas as pd

data = pd.read_csv ('work.csv')   
ndf = pd.DataFrame(data)

print(ndf)
```

```python
import mysql.connector
from mysql.connector import Error
try:
    conn = mysql.connector.connect(host="127.0.0.1",user="root",port ="3306",password="Rishi@123",database="work")
    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("select database();")
        record = cursor.fetchone()
        print("You're connected to database: ", record)
        cursor.execute('DROP TABLE IF EXISTS webscrapping;')
        print('Creating table....')
# in the below line please pass the create table statement which you want #to create
        cursor.execute("CREATE TABLE webscrapping (retailer varchar(30),title varchar(100),product varchar(150),rice varchar(50),available varchar(200),sku varchar(20),website varchar(150))")
        print("Table is created....")
        #loop through the data frame
        for i,row in ndf.iterrows():
            #here %S means string values 
            sql = "INSERT INTO work.webscrapping VALUES (%s,%s,%s,%s,%s,%s,%s)"
            cursor.execute(sql, tuple(row))
            print("Record inserted")
            # the connection is not auto committed by default, so we must commit to save our changes
            conn.commit()
except Error as e:
            print("Error while connecting to MySQL", e)
```
